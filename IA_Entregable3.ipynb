## **Revision de las columnas inpactantes para el analisis**

```python
#Verificando Valores unicos en cada una de las Columnas
for col in bdAurelion.columns:
  print('-' * 60 )
  print(f"Valores unicos en la columna {col}: {bdAurelion[col].nunique()}")
  if bdAurelion[col].nunique() < 10:
    print(f"Valores: {list(bdAurelion[col].unique())}")
    print('=' * 60)
```

## Usando Modelo K-means ##

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.datasets import make_blobs
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8')
sns.set_palette('viridis')

```

```python
print(f"\nTotal clientes: {len(bdAurelion)}")

bdAurelion.head()
```

**Visualizacion Exploratoria**

```python
plt.figure(figsize=(14,8))
sns.scatterplot(data=bdAurelion, x='importe', y='cantidad', alpha=0.6)
plt.title('Importe Total vs Cantidad Vendida')
plt.xlabel('Importe (S/.)')
plt.ylabel('Cantidad')

plt.show()
```

```python
bdAurelion['Categoria'] = bdAurelion['Categoria'].map({'Limpieza': 1, 'Alimentos': 2})
bdAurelion['Categoria'] = bdAurelion['Categoria'].astype(int)
bdAurelion.head()

bdAurelion1 = bdAurelion[['Categoria', 'cantidad', 'precio_unitario', 'importe']]
bdAurelion1.head()
```

```python
scaler = StandardScaler() # z-score, esto optimiza el modelo
data_scaled = scaler.fit_transform(bdAurelion1)

df = pd.DataFrame(data_scaled, columns=bdAurelion1.columns).head().round(2)
df
```

```python
inertia = []
k_range = range(2, 8)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(k_range, inertia, marker='o')
plt.title('MÃ©todo del Codo - Â¿CuÃ¡ntos grupos?')
plt.xlabel('NÃºmero de clusters')
plt.ylabel('Inercia')
plt.axvline(3, color='purple', linestyle='-', label='K=3')
plt.legend()
plt.show()
```

## Aplicando KMens ##

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data_scaled)

bdAurelion1['cluster'] = kmeans.labels_

print("Clientes por grupo:")
print()
print(bdAurelion1['cluster'].value_counts().sort_index())
```

##Visualizar Grupo

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=bdAurelion1, x='importe', y='cantidad', hue='cluster', palette='Set2')
plt.title('Clientes Agrupados por K-Means')
plt.xlabel('Importe (S/.)')
plt.ylabel('Cantidad Vendida')
plt.show()
```

## InterpretaciÃ³n de los Grupos ###

```python
print("Promedio por grupo:")
print()
df = bdAurelion1.groupby('cluster').mean().round(0)
df
```

```python
print(f"Silhouette: {silhouette_score(data_scaled, bdAurelion1['cluster']):.3f}")
print(f"Calinski-Harabasz: {calinski_harabasz_score(data_scaled, bdAurelion1['cluster']):.0f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8')
sns.set_palette("viridis")
```

## Preparar datos de entrenamiento y pruebas

```python
X =bdAurelion1[['cantidad', 'precio_unitario', 'Categoria']] # Variables predictoras o independientes
y = bdAurelion1['importe'] # Variable objetivo

# Estandarizar (opcional, pero mejora interpretaciÃ³n)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividimos datos de entrenamiento y test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42) # test_size es el tamaÃ±o del test
```

## VisualizaciÃ³n de trains/test ##

```python
# Crear el scatter plot
plt.figure(figsize=(10, 7))

# Puntos de ENTRENAMIENTO (azules, cÃ­rculo)
plt.scatter(X_train[:, 0], y_train,
            color='blue', label='Entrenamiento', alpha=0.7, s=60, edgecolors='k')

# Puntos de PRUEBA (rojos, triÃ¡ngulo)
plt.scatter(X_test[:, 0], y_test,
            color='red', label='Prueba', marker='^', alpha=0.8, s=80, edgecolors='k')

# Etiquetas y tÃ­tulo
plt.xlabel('Importe Total')
plt.ylabel('Cantidad Vendida')
plt.title('Conjuntos de Entrenamiento y Prueba (superpuestos)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Entrenar Modelo ##

```python
modelo = LinearRegression()
modelo.fit(X_train, y_train)

print()
print("Coeficientes:")
print("-------------")
for var, coef in zip(['cantidad', 'precio_unitario', 'Categoria'], modelo.coef_):
    print(f"  {var}: {coef:,.1f}")

print()
print(f"Intercepto: {modelo.intercept_:,.0f}")
```

## Prueba del Modelo ##

```python
y_pred = modelo.predict(X_test)
```

## Evaluar el modelo ##

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("=== MÃ‰TRICAS DE REGRESIÃ“N ===")
print(f"MAE  (Error absoluto medio). En promedio se equivoca por: ${mae:,.0f}")
print(f"MSE  (Error cuadrÃ¡tico medio). Castiga mÃ¡s los errores grandes â†’ Ãºtil para optimizar: ${mse:,.0f}")
print(f"RMSE (RaÃ­z del error cuadrÃ¡tico). En promedio se equivoca por: ${rmse:,.0f} (penaliza errores grandes)")
print(f"RÂ²   (CuÃ¡nto explica el modelo en % la variaciÃ³n del precio): {r2:.3f} â†’ {r2*100:.1f}%")

# GrÃ¡fico real vs predicho
plt.figure(figsize=(14,7))
plt.scatter(y_test, y_pred, alpha=0.6, color='teal')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel("Importe Real")
plt.ylabel("Importe Predicho")
plt.title("Real vs Predicho (LÃ­nea roja = perfecto)")
plt.grid(True, alpha=0.3)
plt.show()
```
